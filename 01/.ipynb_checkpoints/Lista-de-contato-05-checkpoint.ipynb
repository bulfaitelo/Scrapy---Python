```python
def criarContato():
    print("Digite as informações do contato")
    nome = input("nome: ")
    email = input('E-mail: ')
    telefone = input("Telefone: ")
    return {
        'nome': nome, 
        'email': email,
        'telefone':telefone
    }
```

```python
def salvarContato(nome, email, telefone):
    arquivo = open('contatos.csv', 'a')
    arquivo.write('%s, %s, %s\n' % (nome, email, telefone))
    arquivo.flush()
    arquivo.close()
```
