```python
import random
```

```python
continua = True 
while continua:
    faces = input("Digite o numero de faces(ou 'parar' para sair): ")
    if faces == "parar":
        continua = False
    else:
        faces = int(faces)
        dados = input('Digite o numero de Dados: ')
        for i in range(int(dados)):
            numero = random.randint(1,faces)
            print("Seu numero é: %s" % numero)
print("Finalizou o programa")
```
