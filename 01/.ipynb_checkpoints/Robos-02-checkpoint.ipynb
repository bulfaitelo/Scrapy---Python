# Robos
a matriz vai de 0x0 a 10x10

```python
class Robo(object):
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def move_up(self):
        if self.y < 10:
            self.y = self.y + 1
        else:
            print('Movimento proibido')
        
    def move_down(self):
        self.y = self.y - 1
        
    def move_left(self):
        self.x = self.x - 1
        
    def move_right(self):
        self.x = self.x + 1
```

```python
r1 = Robo(5,5)
```

```python
r1.move_down
```
